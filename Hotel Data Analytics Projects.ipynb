```python
avg, std = df_bookings.revenue_generated.mean(), df_bookings.revenue_generated.std()
```

```python
higher_limit = avg + 3*std
higher_limit
```

```python
lower_limit = avg - 3*std
lower_limit
```

```python
df_bookings[df_bookings.revenue_generated<=0] #removing outliers in revenue genrated column
```

```python
df_bookings[df_bookings.revenue_generated>higher_limit]
```

### ==> 3. Data Transformation
Data Transformation is your creating new column you are aggregating data. there could be variety of data transformation

**Create occupancy percentage column**

```python
df_agg_bookings.head()
```

```python
df_agg_bookings['occ_pct'] = df_agg_bookings["occ_pct"].apply(lambda x : round(x*100,2))
df_agg_bookings.head(5)
```

***
### ==> 4. Insights Generation
***

**1. What is an average occupancy rate in each of the room categories?**

```python
df_agg_bookings.head(3)
```

```python
df_agg_bookings.groupby("room_category")["occ_pct"].mean().round(2) #here we can get average occupancy rate in each of room category
```

I don't understand RT1, RT2 etc. Print room categories such as Standard, Premium, Elite etc along with average occupancy percentage so here we are doing merege with df_rooms to get room_class

```python
df = pd.merge(df_agg_bookings, df_rooms, left_on="room_category", right_on="room_id")
df.head(4)
```

```python
df.drop("room_id",axis=1, inplace=True) #here we are droping the room_id column
df.head(4)
```

```python
df.groupby("room_class")["occ_pct"].mean().round(2)
```

**2. Print average occupancy rate per city**

```python
df_hotels.head(3)
```

```python
df = pd.merge(df, df_hotels, on="property_id") #here again we are merging using property id
df.head(3)
```

```python
df.groupby("city")["occ_pct"].mean().plot(kind="bar")
```

**3. When was the occupancy better? Weekday or Weekend?**

```python
df.head(3)
```

```python
df = pd.merge(df, df_date, left_on="check_in_date", right_on="date") #here we are again merging to get day_type
df.head(3)
```

```python
df.groupby("day_type")["occ_pct"].mean().round(2) #checking when is the occupancy rate is better
```

**4: In the month of June, what is the occupancy for different cities**

```python
df["mmm yy"].unique()
```

```python
df_june_22 = df[df["mmm yy"]=="Jun 22"] #here we are filtering data on june column
df_june_22.head(4)
```

```python
df_june_22.groupby('city')['occ_pct'].mean().round(2).sort_values(ascending=False) # grouping june data basis of cities
```

so here we get to know that dehli is performing well in june month & in banglore occupacny rate is lowest

```python
df_june_22.groupby('city')['occ_pct'].mean().round(2).sort_values(ascending=False).plot(kind="bar")
```
